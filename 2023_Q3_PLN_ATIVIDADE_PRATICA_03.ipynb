#### **INTRODUÇÃO**

**Descrição da API**

O YouTube Data API é uma ferramenta desenvolvida pelo Google, que oferece acesso a uma ampla gama de informações e funcionalidades para o YouTube. É portanto uma ferramenta indispensável para desenvolvedores que desejam criar aplicativos, análises ou integrações relacionadas ao YouTube.

Esta API permite que os desenvolvedores acessem informações detalhadas sobre vídeos, canais, listas de reprodução, comentários entre outras informações, o que é essencial para criar aplicativos que ofereçam funcionalidades avançadas, como ferramentas de pesquisa, gerenciamento de conteúdo ou análise de dados.

Além disso, e de particular interesse para um trabalho de PLN, através desta API podemos coleta de transcrições de vídeos, legendas, descrições e comentários. Esses dados contêm uma riqueza de texto que pode ser usado para treinar modelos de PLN, melhorar a compreensão de linguagem natural e desenvolver aplicações como análise de sentimentos e resumo de text. Mais ainda, podemos utilizá-la para identificar tópicos e tendências com essas informações criar modelos de recomendação.

#### **EXEMPLOS PRÁTICOS**

Para o exemplo prático, vamos mostrar como obter a lista de comentários de um vídeo. A partir desta lista, podemos realizar diversos processamentos e obter informações relevantes, como sentimentos gerais e tópicos frequentes. Por simplicidade, mostraremos como obter uma lista com as palavras mais frequentes e uma seção de comentários.

```
# Recebendo uma lista dos 20 comentários mais relevantes em um vídeo
request = youtube.commentThreads().list(
    part="snippet",
    videoId="GUQWQVMEyZA",
    order="relevance",
)

commentsResponse = request.execute()
```

```
print(json.dumps(commentsResponse, indent=4))
```

```
comments = [commentElement['snippet']['topLevelComment']['snippet']['textOriginal'] for commentElement in commentsResponse['items']]

for comment in comments:
  print(comment)
  print('-----------------------------')
```

```
# Palavras mais usadas
import re

words = dict()
for comment in comments:
  matches = re.findall('[a-zA-Z\']+', comment)
  for mtch in matches:
    word = mtch.lower()
    try:
      words[word] = words[word] + 1
    except:
      words[word] = 1


print(dict(sorted(words.items(), reverse=True, key=lambda item: item[1])))
```

Uma "word cloud" (nuvem de palavras) é uma representação visual de um conjunto de palavras em que o tamanho de cada palavra é proporcional à sua frequência no texto original, isto é, as palavras mais frequentes são exibidas em um tamanho maior, enquanto as menos frequentes são menores. É comum remover as "stopwords" (palavras de ligação e muito frequentes, como "the", "and" e "in" em inglês) antes de criar uma word cloud, pois essas palavras não carregam um significado substancial e podem distorcer a representação visual. Vejamos abaixo como criar uma word cloud:

```
# Gerando uma word cloud

import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
import random

# Baixamos a base de dados de stopwords em inglês
nltk.download('stopwords')

stop_words = set(stopwords.words("english"))

filtered_words = dict((word.capitalize(),count) for word, count in words.items() if word not in stop_words)

# Criamos a word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").fit_words(filtered_words)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.show()
```
